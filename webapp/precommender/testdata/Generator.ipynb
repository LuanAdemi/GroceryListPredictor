```python
from generator import generate, generateFromGenFile
import matplotlib.pyplot as plt
```

```python
prods = ["APFELMUSS", "BATTERIEN", "BITTERLEMON"]
probs = [0.06,0.2,0.8]
data, pxs = generate(probs,prods)
f, axes = plt.subplots(3,1, figsize=(9, 4))
for i in range(len(axes)):
    
    axes[i].plot(data[:,i], c='r')
    axes[i].plot(pxs[:,i], c='g')
f.legend(['Time series', 'Probability'], bbox_to_anchor=(0., 0, 1, .102), loc='lower left',
           ncol=2, mode="expand", borderaxespad=0.)
```

```python
for i in range(1, len(data)+1):
    f = open("gu_" + str(i) + ".txt", 'w')
    f.write(str(features[0]))
    for feature in features[1:]:
        f.write(",")
        f.write(feature)
    for d in data[i-1][1:]:
        f.write("\n")
        f.write(str(d[0]))
        for feature in d[1:]:
            f.write(",")
            f.write(str(int(feature)))
    f.close()
```
